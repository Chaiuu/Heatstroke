**Model Building (Ridge Regression)**

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error


data = df

# Identify categorical and numerical columns
categorical_cols = data.select_dtypes(include=['object', 'bool']).columns.tolist()
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_cols.remove('hrv_rmssd')  # Exclude the target variable from the numerical columns

# Define preprocessing for numerical columns: impute missing values and scale
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())])

# Define preprocessing for categorical columns: impute missing values and apply one-hot encoding
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)])

# Preprocess the dataset
X = data.drop('hrv_rmssd', axis=1)
y = data['hrv_rmssd']
X_preprocessed = preprocessor.fit_transform(X)

# Split the dataset into training and testing set
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42)

# Initialize and train the Ridge regression model
ridge_model = Ridge(alpha=1.0, random_state=42)
ridge_model.fit(X_train, y_train)

# Predict on the testing set
y_pred = ridge_model.predict(X_test)

# Calculate and print the mean squared error and root mean squared error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
```

**HRV SD1 (Standard Deviation 1):**

**Description:** This metric is one of the components derived from Poincaré plot analysis, which is a scatter plot of current RR intervals against the next consecutive RR intervals (RR intervals are the time intervals between consecutive heartbeats). HRV SD1 measures the standard deviation of the distances perpendicular to the line of identity (line at 45 degrees in the plot). It is often used to assess short-term variations in heart rate.

**HRV sdsd (Successive Differences Standard Deviation):**

**Description:** This metric calculates the standard deviation of successive differences between adjacent RR intervals. It is used to evaluate the variability in the short-term intervals between heartbeats, providing insights into the autonomic nervous system's parasympathetic function.
HRV cvsd (Coefficient of Variation of Successive Differences):
Description: Similar to sdsd, this measure is used to examine the variability in heartbeat intervals. However, it normalizes the standard deviation by the mean of the RR intervals, offering a relative measure of variability irrespective of the heart rate.

**HRV sdnn (Standard Deviation of NN Intervals):**

**Description:** This is the standard deviation of all NN intervals (normal-to-normal intervals, which are intervals between adjacent QRS complexes resulting from sinus node depolarizations). HRV sdnn is a basic measure that reflects all cyclic components responsible for variability in the period of recording, representing overall HRV.

**HRV std (Standard Deviation):**

**Description:** This could refer to the standard deviation of a specific subset of data or the overall HRV data being analyzed. It quantifies the dispersion or variability from the average, and in HRV context, it often mirrors HRV sdnn unless specified differently.

```python
import numpy as np

def calculate_hrv_metrics(rr_intervals):
    """
    Calculate HRV metrics from RR intervals.

    Args:
    rr_intervals (list): A list of RR intervals in milliseconds.

    Returns:
    dict: A dictionary containing various HRV metrics.
    """
    # Convert list of intervals to numpy array for computation
    rr_array = np.array(rr_intervals)

    # Calculate mean of RR intervals
    mean_nn = np.mean(rr_array)

    # Calculate standard deviation (SDNN)
    sdnn = np.std(rr_array, ddof=1)  # ddof=1 for sample standard deviation

    # Calculate RMSSD
    diff_nn = np.diff(rr_array)
    sq_diff_nn = diff_nn ** 2
    rmssd = np.sqrt(np.mean(sq_diff_nn))

    # Calculate NN50 and pNN50
    nn50 = sum(np.abs(diff_nn) > 50)
    pnn50 = (nn50 / len(diff_nn)) * 100

    # Return a dictionary of HRV metrics
    hrv_metrics = {
        'Mean NN': mean_nn,
        'SDNN': sdnn,
        'RMSSD': rmssd,
        'NN50': nn50,
        'pNN50': pnn50
    }

    return hrv_metrics

# Example usage:
rr_intervals = [800, 830, 810, 840, 790, 810]  # Example RR intervals in milliseconds
hrv_metrics = calculate_hrv_metrics(rr_intervals)
print(hrv_metrics)
```

```python
import joblib  # Used for loading a pre-trained model

def predict_heat_stress(hrv_metrics):
    """
    Predict the heat stress level based on HRV metrics using a pre-trained model.

    Args:
    hrv_metrics (dict): A dictionary containing HRV metrics.

    Returns:
    float: The predicted heat stress level.
    """
    # Load your trained model (ensure the model is in the same directory or provide the correct path)
    model = joblib.load('model.pkl')

    # Prepare the input data for the model (model expects an array of features)
    # Assuming the model expects features in the order: ['Mean NN', 'SDNN', 'RMSSD', 'NN50', 'pNN50']
    input_features = np.array([[hrv_metrics['Mean NN'], hrv_metrics['SDNN'], hrv_metrics['RMSSD'], hrv_metrics['NN50'], hrv_metrics['pNN50']]])

    # Predict using the model
    prediction = model.predict(input_features)[0]  # [0] to extract single prediction value

    return prediction
```

```python
import numpy as np

def calculate_hrv_metrics(rr_intervals):
    rr_intervals = np.array(rr_intervals)  # Convert list to numpy array for easier calculations

    # Calculate mean of RR intervals
    mean_rr = np.mean(rr_intervals)

    # Calculate max and min RR intervals
    max_rr = np.max(rr_intervals)
    min_rr = np.min(rr_intervals)

    # Calculate HRV metrics such as RMSSD (Root Mean Square of Successive Differences)
    differences = np.diff(rr_intervals)  # Calculate successive differences
    squared_differences = np.square(differences)
    rmssd = np.sqrt(np.mean(squared_differences))

    return mean_rr, max_rr, min_rr, rmssd

# Example usage
rr_intervals = [800, 1000, 950, 880, 760]  # RR intervals in milliseconds
mean_rr, max_rr, min_rr, rmssd = calculate_hrv_metrics(rr_intervals)
print(f"Mean RR: {mean_rr} ms, Max RR: {max_rr} ms, Min RR: {min_rr} ms, RMSSD: {rmssd} ms")
```
